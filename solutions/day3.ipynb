```python
# filename = "AoC day 3 example data.txt"
filename = "AoC day 3 data.txt"
with open(filename, 'r') as f:
    data = [line.strip() for line in f.readlines()]
```

```python
priority = {}
for n, i in enumerate(range(97, 123), 1):
    priority[chr(i)] = n
for n, i in enumerate(range(65, 91), 27):
    priority[chr(i)] = n
```

```python
summed_priority = 0
for line in data:
    n_items = len(line)
    s = n_items // 2
    first_compartment = list(line[:s])
    second_compartment = list(line[s:])
    in_both = set.intersection(set(first_compartment), set(second_compartment)).pop()
    summed_priority += priority[in_both]
print(summed_priority)
```

```python
grouped_data = list(zip(data[::3], data[1::3], data[2::3]))
grouped_data
```

```python
summed_priority = 0
for gd in grouped_data:
    summed_priority += priority[set.intersection(*[set(l) for l in gd]).pop()]
print(summed_priority)
```
